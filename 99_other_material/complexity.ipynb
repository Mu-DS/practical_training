```python
import matplotlib.pyplot as plt
import numpy as np

vals = [1, 5, 100, 10000, 1000000]
vals = np.log10(vals)
vals = [1, 2, 3, 4, 5]

fig, ax = plt.subplots(1, 1, figsize=(12, 7.2))
ax.xaxis.set_ticks(vals)
ax.xaxis.set_ticklabels([
    "1 person (you)",
    "5 people (team)",
    "100 people (group)",
    "10,000 people (division)",
    "1,000,000 people (world)"])
ax.yaxis.set_ticks(vals)
ax.yaxis.set_ticklabels([
    "1 day",
    "1 week",
    "3 months",
    "1 year",
    "10 years"])
```
